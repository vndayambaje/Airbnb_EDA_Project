## Looking at the average price accross countries 

```python
# Group by country and calculate the average price
country_avg_price = df.groupby('country')['price'].mean().reset_index()

# Rename columns for clarity
country_avg_price.columns = ['Country', 'Average Price']

# Round average prices to one decimal place
country_avg_price['Average Price'] = country_avg_price['Average Price'].round(1)

# Sort the results in descending order
country_avg_price = country_avg_price.sort_values(by='Average Price', ascending=False)

# Display the result
print(country_avg_price)
```

## 3. Which countries have the highest and lowest average ratings?


#### First looking for how many ratings are listed as a string "New" Vs The ratings that are Numeric

```python


# Convert 'rating' to string to handle mixed types uniformly
df['rating'] = df['rating'].astype(str)

# Count occurrences of 'New'
new_count = df['rating'].str.contains('New', na=False).sum()

# Filter out 'New' values to focus on numeric and NaN values
non_new_ratings = df[~df['rating'].str.contains('New', na=False)]

# Convert remaining ratings to numeric, coercing errors to NaN
non_new_ratings['rating'] = pd.to_numeric(non_new_ratings['rating'], errors='coerce')

# Drop rows with NaN values in 'rating'
numeric_ratings = non_new_ratings.dropna(subset=['rating'])

# Count occurrences of numeric ratings
numeric_counts = numeric_ratings['rating'].value_counts()

# Count occurrences of NaN values
nan_count = df['rating'].isna().sum()

# Display results
print(f"Number of 'New' ratings: {new_count}")
print("\nCounts of numeric ratings:")
print(numeric_counts)
print(f"\nNumber of NaN ratings: {nan_count}")
```

#### Changing all "New" and "NaN" values in ratings to "0" rating

```python
# Convert 'rating' to string to handle mixed types uniformly
df['rating'] = df['rating'].astype(str)

# Replace 'New' with '0'
df['rating'] = df['rating'].replace('New', '0')

# Convert 'rating' to numeric (float), coercing errors to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Fill NaN values with 0
df['rating'] = df['rating'].fillna(0)

# Display updated DataFrame
print("Updated DataFrame:")
print(df.head())

# Check the value counts for the updated ratings
rating_counts = df['rating'].value_counts()

# Display the result
print("\nCounts of updated ratings:")
print(rating_counts)
```

### Now We can get the countries with highest and lowest ranking 

```python

# Ensure 'rating' is numeric, force errors to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows where 'rating' is NaN
df = df.dropna(subset=['rating'])

# Convert 'rating' column to float
df['rating'] = df['rating'].astype(float)

# Group by country and calculate the average rating
country_avg_rating = df.groupby('country')['rating'].mean().reset_index()

# Rename columns for clarity
country_avg_rating.columns = ['Country', 'Average Rating']

# Round average ratings to one decimal place
country_avg_rating['Average Rating'] = country_avg_rating['Average Rating'].round(1)

# Sort the results in descending order
country_avg_rating = country_avg_rating.sort_values(by='Average Rating', ascending=False)

# Display the result
print(country_avg_rating)
```

## 5. What are the most common amenities offered in listings in each country?

```python
from collections import Counter

# Split amenities into individual items
df['amenities'] = df['amenities'].apply(lambda x: x.replace('[', '').replace(']', '').replace('"', '').split(','))

# Count frequency of each amenity per country
def count_amenities(amenities_list):
    return Counter([amenity.strip() for sublist in amenities_list for amenity in sublist])

country_amenities = df.groupby('country')['amenities'].apply(list).apply(count_amenities).reset_index()
print(country_amenities)
```

## 6. Which countries have the most diverse range of amenities provided by Airbnb listings?

```python


# Ensure the 'amenities' column is treated as a string
df['amenities'] = df['amenities'].astype(str)

# Split the amenities into lists based on a common delimiter (e.g., comma)
df['amenities_list'] = df['amenities'].apply(lambda x: x.split(','))

# Create a DataFrame where each row contains a country and an individual amenity
amenities_expanded = df.explode('amenities_list')

# Remove any leading or trailing whitespace from amenities
amenities_expanded['amenities_list'] = amenities_expanded['amenities_list'].str.strip()

# Group by country and count unique amenities
country_amenity_diversity = amenities_expanded.groupby('country')['amenities_list'].nunique().reset_index()

# Rename columns for clarity
country_amenity_diversity.columns = ['Country', 'Unique Amenities Count']

# Sort results by the number of unique amenities in descending order
country_amenity_diversity = country_amenity_diversity.sort_values(by='Unique Amenities Count', ascending=False)

# Display the results
print("Countries with the most diverse range of amenities:")
print(country_amenity_diversity)

```

## 7. Which countries have the highest proportion of highly-rated listings?

```python
# Define a threshold for high ratings
high_rating_threshold = 4.5

# Calculate proportion of listings with high ratings in each country
high_rated_listings = df[df['rating'] >= high_rating_threshold]
country_high_ratings = high_rated_listings.groupby('country').size() / df.groupby('country').size()
country_high_ratings = country_high_ratings.reset_index(name='proportion_high_ratings')

# Round proportions to one decimal place
country_high_ratings['proportion_high_ratings'] = country_high_ratings['proportion_high_ratings'].round(1)

# Sort the results in descending order
country_high_ratings = country_high_ratings.sort_values(by='proportion_high_ratings', ascending=False)

# Display the result
print(country_high_ratings)
```

## 8. How does the number of listings managed by hosts vary across countries?

```python
# Count the number of listings per host
host_listing_counts = df.groupby(['country', 'host_id']).size().reset_index(name='num_listings')

# Summarize the information per country
country_host_listing_counts = host_listing_counts.groupby('country')['num_listings'].describe().reset_index()

# Display the result
print(country_host_listing_counts)
```

## 9. In which countries is there the most significant potential for Airbnb to invest based on current business success rates?

```python


# Convert 'rating' and 'reviews' to numeric, coercing errors to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['reviews'] = pd.to_numeric(df['reviews'], errors='coerce')

# Define business success criteria
success_criteria = (df['rating'] >= 4.5) & (df['reviews'] >= 50)

# Calculate the proportion of listings meeting success criteria in each country
successful_listings = df[success_criteria]
country_success_rate = successful_listings['country'].value_counts(normalize=True) * 100

# Sort results in descending order
country_success_rate = country_success_rate.sort_values(ascending=False)

# Display the results
print("Proportion of successful listings by country:")
print(country_success_rate)
```
